(WIKIPEDIA: MultikollinearitÃ¤t ist ein Problem der Regressionsanalyse und liegt vor, wenn zwei oder mehr erklÃ¤rende Variablen eine sehr starke Korrelation miteinander haben. Zum einen wird mit zunehmender MultikollinearitÃ¤t das Verfahren zur SchÃ¤tzung der Regressionskoeffizienten instabil und Aussagen zur SchÃ¤tzung der Regressionskoeffizienten zunehmend ungenau. Zum anderen ist die Modellinterpretation nicht mehr eindeutig. Das klassische Symptom von starker MultikollinearitÃ¤t ist ein hohes BestimmtheitsmaÃŸ einhergehend mit niedrigen t-Werten fÃ¼r die einzelnen Regressionsparameter. // AB: fÃ¼hrt zu SchÃ¤tzfehlern.)

***Verwenden Sie die in der Heatmap angezeigte Korrelationsmatrix, um kollineare Merkmale auszuwÃ¤hlen und zu entfernen. Denken Sie daran, die Antwortvariablen aus der Matrix auszuschlieÃŸen, um sicherzustellen, dass sie in Ihrem endgÃ¼ltigen Modellentwicklungsdatensatz beibehalten werden. WÃ¤hlen Sie dann die Features aus, deren Entfernung zu mehr als 95% korreliert ist.***

```python
# 1. exclude response variables, 2. extract upper triangle, 3.remove cols > .95
corr_matrix = dfno.drop(['AdultWeekday', 'AdultWeekend', 'daysOpenLastYear', 'projectedDaysOpen'], axis = 1).corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
```

Nun befassen wir uns mit der Feature-Erstellung dieses Schritts. Wir kÃ¶nnen zusÃ¤tzliche Features durch viele Methoden erstellen, z. B. durch Kombinieren mehrerer Features, Gruppieren von Features in Bins oder sogar durch Anwenden eines unbeaufsichtigten Klassifizierungsalgorithmus, wie z. B. k-means Clustering und Verwenden der Cluster als Features in unserem Modellentwicklungsdatensatz.

Beim Clustering werden im Wesentlichen Muster in Daten gefunden, wenn wir nicht im Voraus wissen, wonach wir suchen. Der K-Means-Algorithmus ist eine MÃ¶glichkeit zum Clustering. Es ordnet unsere Daten in Gruppen (oder "Cluster") ein, die dann zu Merkmalen fÃ¼r die weitere Analyse werden kÃ¶nnen. Was der k-means-Algorithmus ausgibt, hÃ¤ngt jedoch von der Anzahl der Cluster ab, die wir uns vorgenommen haben. Wir werden eine Methode namens "Elbow Plotting" verwenden, um diese Zahl zu bestimmen.

Wir werden Clustering und k-means in der Einheit "UnÃ¼berwachtes Lernen" dieses Kurses ausfÃ¼hrlich behandeln, aber im Moment werden wir es nur fÃ¼r die Erstellung von Features verwenden, was ein wesentlicher Bestandteil der exploratorischen Datenanalyse ist. Machen Sie sich keine Sorgen, wenn Sie die Details noch nicht verstehen.

FÃ¼hren Sie die nÃ¤chsten beiden Zellen unten aus, um ein Ellbogen-Diagramm zu erstellen. Das Elbow-Diagramm ist ein Diagnosetool, mit dem wir die Anzahl der Cluster bestimmen kÃ¶nnen, die in unsere k-means-Clustering-Implementierung einbezogen werden sollen.

In diesem Beispiel wird der Fehler zwischen Clustern und innerhalb von Clustern fÃ¼r einen Bereich von 1 bis 11 Clustern verglichen, und es scheint, dass der Ellbogen zwischen zwei und vier liegt. Daher setzen wir den Parameter k = 3.

```python
from sklearn.cluster import KMeans
```

```python
a = dfno.drop(['Name','state'], axis =1).values
a
```

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(a)
    kmeans.fit(a)
    Error.append(kmeans.inertia_)
import matplotlib.pyplot as plt
plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

#### Fit the kmeans algorithm with the k parameter = 3 and plot the results

```python
kmeans3 = KMeans(n_clusters=3)
y_kmeans3 = kmeans3.fit_predict(a)
plt.scatter(a[:, 0], a[:, 1], c=y_kmeans3, s=50, cmap='viridis')

centers = kmeans3.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
def df_copy(df):
    kopie = df.copy()
    return kopie.to_csv("D:\\BoOst\\01_DS\\Projects\\SB_05_GC-1_MountainSki\\data\\kopie_reduced_0724.csv")

def get_rid_high_ccfeats(df):
    # step 1 drop column with correlation feature > 95%
    corr_matrix = df.drop(['AdultWeekday', 'AdultWeekend', 'daysOpenLastYear', 'projectedDaysOpen'], axis = 1).corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
    to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
    df_copy(df)
    df.drop(df[to_drop], axis=1, inplace=True)
    return to_drop, df.shape

get_rid_high_ccfeats(df_iqr_remo)
```

```python
def elbow_plot(df):
    from sklearn.cluster import KMeans
    x = df.drop(['Name','state'], axis =1).values

    Error =[]
    for i in range(1, 11):
        kmeans = KMeans(n_clusters = i).fit(x)
        kmeans.fit(x)
        Error.append(kmeans.inertia_)
    
    import matplotlib.pyplot as plt
    plt.plot(range(1, 11), Error)
    plt.title('Elbow method')
    plt.xlabel('No of clusters')
    plt.ylabel('Error')
    plt.show()
    
elbow_plot(df_iqr_remo)
```

```python
def kmeans_col(df, clusterint):
    from sklearn.cluster import KMeans
    x = df.drop(['Name','state'], axis =1).values
    kmeans0 = KMeans(n_clusters=clusterint)
    y_kmeans0 = kmeans0.fit_predict(x)
    df.loc[:,"clusters"] = y_kmeans0
    
    import matplotlib.pyplot as plt
    plt.scatter(x[:, 0], x[:, 1], c=y_kmeans0, s=50, cmap='viridis')
    centers = kmeans0.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5);
    return df.clusters

kmeans_col(df_iqr_remo,3) 
```
